# 1.- Identificación del problema

Bob ha comenzado su propia empresa de telefonía móvil.

No sabe cómo estimar el precio de los móviles que fabrica su empresa. En este competitivo mercado de telefonía móvil no se puede simplemente asumir cosas. Para resolver este problema, recopila datos de ventas de teléfonos móviles de varias empresas.

Bob quiere averiguar alguna relación entre las funciones de un teléfono móvil (p. ej., RAM, memoria interna, etc.) y su precio de venta.

En este problema, no tiene que predecir el precio real, sino un rango de precios que indica qué tan alto es el precio

**OBJETIVO:** Predecir en qué rango (0,1,2,3) oscilan los precios de los telefonos moviles en el mercado dadas sus caracteristicas (train) y luego estimar el precio de los moviles que produce BOB (test)

# 4.- Preparando datos para el modelado

```python
pcel_train_ = pcel.sample(frac= .80)
pcel_test = pcel.drop(pcel_train_.index)
```

```python
#Dividimos nuesros datos train, validation and test
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
x = pcel_train_.drop('price_range', axis = 1)
y =pcel_train_['price_range']

x_test = pcel_test.drop('price_range', axis = 1)
y_test = pcel_test['price_range']

x_train, x_valid, y_train, y_valid = train_test_split(x, y, random_state=21, test_size = 0.3)
```

# 5.- Random Forest 🌳🌲🌴    🦕

```python
from sklearn.ensemble import RandomForestClassifier

clf_rf = RandomForestClassifier(random_state = 0)
clf_rf.fit(x_train, y_train)
```

```python
clf_rf.get_params()
```

```python
print("Puntaje de entrenamiento (train score):", clf_rf.score(x_train,y_train))
print("Puntaje de validación (validation score):", clf_rf.score(x_valid,y_valid))
print("Puntaje de evaluación (test score):", clf_rf.score(x_test,y_test))
```

## 5.1.- Grid Search

```python
from sklearn.model_selection import GridSearchCV
clf_rf_ = RandomForestClassifier(random_state = 0)
model_params = {
    'max_depth': [5,10,15,20,50],
    'n_estimators':[100,200,300,500],
    'min_samples_split': [0,2,5,10,20],
}

clf_grid = GridSearchCV(clf_rf_,model_params, cv = 2, verbose=10)
clf_grid.fit(x_train,y_train)
```

```python
clf_grid.best_estimator_.get_params()
```

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import f1_score

clf_rf_best = RandomForestClassifier(bootstrap= True,
 ccp_alpha= 0.0,
 class_weight= None,
 criterion= 'gini',
 max_depth= 15,
 max_features= 'auto',
 max_leaf_nodes= None,
 max_samples= None,
 min_impurity_decrease= 0.0,
 min_samples_leaf= 1,
 min_samples_split= 2,
 min_weight_fraction_leaf= 0.0,
 n_estimators= 200,
 n_jobs= None,
 oob_score= False,
 random_state= 0,
 verbose= 0,
 warm_start= False)
clf_rf_best.fit(x_train,y_train)

#Imprimimos Scores Acuracy
print("Train Score Grid Search (Accuracy):", clf_rf_best.score(x_train,y_train))
print("Validation Grid Search Score (Accuracy):", clf_rf_best.score(x_valid,y_valid))
print("Test Score Grid Search (Accuracy):", clf_rf_best.score(x_test,y_test))

z=pcelt.drop(['id'], axis=1)
#Imprimimos Scores F1-Score
y_pred_val = clf_rf_best.predict(x_valid)
y_pred_test = clf_rf_best.predict(x_test)
y_pred_def= clf_rf_best.predict(z)
#f1_score(y_test, y_pred_log_reg)

print("Validation Grid Search Score (F1-Score):", f1_score(y_valid, y_pred_val, average='micro'))
print("Test Grid Search Score(F1-Score):", f1_score(y_test,y_pred_test,average='micro'))
print("predicción para el conjunto test que pidió BOB:\n", y_pred_def)
```

# 7.- KNN 🏡 

```python
x_train, x_valid, y_train, y_valid = train_test_split(x, y, random_state=21, test_size = 0.3)
```

```python
#importo el comando para knn y establezco que ensaye con hasta 11 vecinos
from sklearn.neighbors import KNeighborsClassifier
scores=[]
for n_neighbors in range(1,11):
  knn = KNeighborsClassifier(n_neighbors = n_neighbors)
  knn.fit(x_train,y_train)
  scores.append(knn.score(x_valid, y_valid))
```

```python
#graficando para ver cuantos vecinos manejar
plt.plot(scores)
plt.title("relación entre ajuste y cantidad de vecinos")
plt.xlabel("knn")
plt.ylabel("Nivel de ajuste")
print(scores)
```

```python
#Ajustando el knn
knn_final = KNeighborsClassifier(n_neighbors = 4)
knn_final.fit(x,y)

#los resultados hacen que crea que es perfecto pero lo dudo
print("entrenamiento:",knn_final.score(x_train, y_train))
print("validación:",knn_final.score(x_valid, y_valid))
print("prueba:", knn_final.score(x_test, y_test))
```
